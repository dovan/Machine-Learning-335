# Gradient Descent

Input data: x=[1,2,4], y=[2,4,5]

Model: slope-only linear regression model

## Defining Cost Function

```python
def J(x) : # Cost Function
    J = 7*x**2 - 20*x + 15
    return J

def dJ(x) : # Cost Function derivative (gradient)
    dJ = 14*x - 20
    return dJ

def x_new(x,alpha) : # update x value 
    x_new = x - alpha * dJ(x)
    return x_new
```

## Function for Gradient Descent Steps

```python
# Gradient Descent Iterations
def gradient_steps(init_x, alpha, iterations) :
    
    #create two subplots for plotting cost function across θ and across iterations
    fig, (ax, bx) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
       
    #initialize cost series
    costs = [] 
    
    #start plotting J(θ) across θ
      
    x_series = np.linspace(0,4,100)        #x value range from 0 to 4 with 100 interval points 
    ax.plot(x_series, J(x_series))         #plot cost function - quadratic function- convex curve
    
    x = init_x               #start plotting 0-estimate starting with initialized value
    ax.plot(x,J(x), 'o', label='x = %.2f' %x )  
   
    #start plotting 0-estimate as we converge towards negative gradient    
    for i in range(iterations):
        cost = J(x)
        costs.append(cost)  
        
        x_old = x
        x = x_new(x,alpha)
            
        ax.plot(x,J(x), 'o', label='x = %.2f' %x) #plot 0-estimate as a dot
        ax.plot([x_old,x],[J(x_old),J(x)], '-')   #plot a line connecting old 0-estimate to new 0-estimate
        
        ax.set_xlabel("θ")
        ax.set_ylabel("J(θ)")
    
    ax.legend()  #add legend with 0-estimate values over the iterations
        
    #start plotting J(θ) across number of iterations    
    
    bx.plot(range(iterations), costs, color='b')  #bx is the second subplot
    bx.set_xlabel("# of iterations")
    bx.set_ylabel("J(θ)")
    
```

## Generating Gradient steps

Input: initial θ value, alpha, number of iterations

```python
gradient_steps( 4, 0.01, 10 )
```

```python
gradient_steps( 4, 0.01, 20 )
```

```python
gradient_steps( 4, 0.15, 20 )
```
