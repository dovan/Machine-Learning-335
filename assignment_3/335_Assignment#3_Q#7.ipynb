```python
# Import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import re

import sklearn
from sklearn.linear_model import LinearRegression

from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# Import data
df = pd.read_csv('data/Life Expectancy Data.csv')


# Drop Null values
df = df.dropna()
df
```

```python
# get first row using row position
print(df.iloc[0])
```

##  Life Expectancy vs. GDP

```python
#create x,y series
x_series = df["GDP"]
y_series = df["Life expectancy "]

#draw scatter plot
plt.scatter(x_series, y_series)
plt.xlabel("GDP")
plt.ylabel("Life Expectancy")

#reshape for modeling
x = x_series.to_numpy().reshape(-1,1)
y = y_series.to_numpy().reshape(-1,1)

#train model
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)

#plot scatterplot
plt.scatter(x,y)

#plot regression_line
plt.plot(x,y_pred, color='b', linewidth=2)

```

## Multivariate Linear Regression

```python
#multivariate regression model with three predictor variables: 'GDP' 'percentage expenditure' 'Schooling'

x_series = df[['GDP','percentage expenditure','Schooling']]
y_series = df["Life expectancy "]


#reshape for modeling
x = x_series.to_numpy().reshape(-1,3)
y = y_series.to_numpy().reshape(-1,1)

#train model
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)

MSE= mean_squared_error(y, y_pred)
R2= r2_score(y, y_pred)

print(model.coef_)
print(model.intercept_)
```
