```python
# Import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import re

import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
# import Housing.csv
df = pd.read_csv('data/diagnosis.csv')
df

#Data preprocessing
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
df
```

# diagnosis vs. perimeter_mean Model

```python
x = df["perimeter_mean"]
y = df["diagnosis"]

x=x.to_numpy().reshape(-1,1)
y=y.to_numpy()

#train logisitc model
logr = LogisticRegression()
logr.fit(x,y)

θ_o=logr.intercept_
θ_1=logr.coef_

#draw logisitc curve using seaborn library
ax=sns.regplot(x="radius_mean", y="diagnosis",  data=df, logistic=True, ci=None, 
            scatter_kws={'color': 'blue'}, line_kws={'color': '#df355a'})

plt.xlabel('Perimeter Mean', fontsize=12)
plt.ylabel('Diagnosis', fontsize=12)

print(θ_o)
print(θ_1)
```

# 2a: Confusion Matrix

```python
y_pred = logr.predict(x)
metrics.confusion_matrix(y,y_pred)
```

# 2c: Precision & Recall with Library Functions

```python
# Precision 
sklearn.metrics.precision_score(y,y_pred)
```

```python
# Verify the Recall 
sklearn.metrics.recall_score(y,y_pred)
```
