```python
#Histograms of a feature
#['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

fig, ax = plt.subplots()
x_index = 2 #picks the third column of features
colors = ['c', 'm', 'y']

for label, color in zip(range(len(iris.target_names)), colors):
    ax.hist(iris.data[iris.target==label, x_index], 
            label=iris.target_names[label],
            color=color)

ax.set_xlabel(iris.feature_names[x_index])
ax.legend(loc='upper right')
plt.show()
```

```python
#Histograms of all four features
#['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

#create four subplots     
fig, ((ax, bx), (cx, dx)) = plt.subplots(nrows=2, ncols=2, figsize=(10,8))

colors = ['c', 'm', 'y']

#first feature: sepal length
x_index = 0
for label, color in zip(range(len(iris.target_names)), colors):
    ax.hist(iris.data[iris.target==label, x_index], 
            label=iris.target_names[label],
            color=color)
ax.set_xlabel(iris.feature_names[x_index])

#second feature: sepal width
x_index = 1
for label, color in zip(range(len(iris.target_names)), colors):
    bx.hist(iris.data[iris.target==label, x_index], 
            label=iris.target_names[label],
            color=color)
bx.set_xlabel(iris.feature_names[x_index])

#third feature: petal length
x_index = 2
for label, color in zip(range(len(iris.target_names)), colors):
    cx.hist(iris.data[iris.target==label, x_index], 
            label=iris.target_names[label],
            color=color)
cx.set_xlabel(iris.feature_names[x_index])

#third feature: petal width
x_index = 3
for label, color in zip(range(len(iris.target_names)), colors):
    dx.hist(iris.data[iris.target==label, x_index], 
            label=iris.target_names[label],
            color=color)
dx.set_xlabel(iris.feature_names[x_index])


ax.legend(loc='upper right')
bx.legend(loc='upper right')
cx.legend(loc='upper right')
dx.legend(loc='upper right')
plt.show()
```

```python
# Scatterplots
#['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']fig, ax = plt.subplots()

#create six subplots using axs is a 2D NumPy array  
fig, axs = plt.subplots(2, 3, figsize=(12,7))

colors = ['c', 'm', 'y']

rows=[0,1]
cols=[0,1,2]

# create feature pairs for scatterplots
xy_pairs=[[0, 1], [0, 2], [0, 3],[1, 2], [1, 3], [2, 3]]
xy_index=0

for i in rows:
    for j in cols: 
        
        x_index = xy_pairs[xy_index][0]
        y_index = xy_pairs[xy_index][1]
        
        xy_index+=1

        for label, color in zip(range(len(iris.target_names)), colors):
            axs[i,j].scatter(iris.data[iris.target==label, x_index], 
                        iris.data[iris.target==label, y_index],
                        label=iris.target_names[label],
                        c=color)

        axs[i,j].set_xlabel(iris.feature_names[x_index])
        axs[i,j].set_ylabel(iris.feature_names[y_index])
        axs[i,j].legend(loc='upper left')


plt.subplots_adjust(wspace=0.2, hspace=0.2) #format subplots 
plt.show()
```
