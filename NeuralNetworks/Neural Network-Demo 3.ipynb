```python
import tensorflow as tf
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from mlxtend.plotting import plot_decision_regions #pip install mlxtend  

#https://www.kaggle.com/code/salehahmedrony/deep-learning-decision-boundary
```

# Load Moon Data

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=1000, noise=0.1, random_state=42)

plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
plt.legend(['0', '1'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Neural Network with no hidden layer

```python
# Define Neural Network model with no hidden layer

model = tf.keras.models.Sequential([tf.keras.layers.Dense(1, input_dim=2, activation='sigmoid')])
model.summary()

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.05), loss='binary_crossentropy', metrics=['accuracy'])

history= model.fit(X_train, y_train, epochs=200, validation_split=0.2, verbose=0)

# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy- no hidden layer')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()
```

```python
#plot decision Boundary

plot_decision_regions(X=X, y=y, clf=model, legend=2)
plt.xlabel("x", size=5)
plt.ylabel("y", size=5)
plt.title('Decision Boundary- no hidden layer', size=10)
plt.show()
```

```python
# Define Neural Network model with one hidden layer

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(2, input_shape=(2,), activation='tanh'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model.summary()

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.05), loss='binary_crossentropy', metrics=['accuracy'])

history= model.fit(X_train, y_train, epochs=45, validation_split=0.2, verbose=0)

# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy- one hidden layer')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()

#plot decision Boundary
plot_decision_regions(X=X, y=y, clf=model, legend=2)
plt.xlabel("x", size=5)
plt.ylabel("y", size=5)
plt.title('Decision Boundary- one hidden layer', size=10)
plt.show()
```
