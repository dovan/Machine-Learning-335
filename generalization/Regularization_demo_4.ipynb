```python
import numpy as np
import pandas as pd
from numpy import random

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore")
```

# Generate Synthetic Polynomial Data

```python
#create x,y arrays
x=np.arange(1,30)

y=[]

for i in range(len(x)):
    sign = random.choice([-1, 1])
    number = random.randint(30)
    y.append(10+ 10*x[i]- 0.18*x[i]**2+sign*number)

y=np.array(y)
#draw scatter plot
plt.scatter(x,y)
```

```python
#linear regression
x = x.reshape(-1,1)
y = y.reshape(-1,1)

#train model
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)

#plot scatterplot
plt.scatter(x,y)

#plot regression_line
plt.plot(x,y_pred, color='m')

mse=mean_squared_error(y, y_pred)
print(mse)

r2= r2_score(y, y_pred)
print(r2)
```

# Polynomial Models

```python
#train polynomial regression models with degrees: 1,2,8

degrees = [1,2,8]
colors=['b','y','m']

plt.scatter(x, y, color='#71b3ef')

for i in range(len(degrees)):  
    poly = PolynomialFeatures(degree=degrees[i])
    poly_features = poly.fit_transform(x)
    
    model = LinearRegression()
    model.fit(poly_features, y)
    
    #predict y-values
    y_pred = model.predict(poly_features)    
    
    #model score
    r2= r2_score(y, y_pred)    
    
    sns.regplot(x=x,y=y_pred,order=degrees[i], scatter=False, line_kws={'color': colors[i], 'linewidth':2}, 
                ci=None, label='poly: '+str(degrees[i])+' Model Score: '+str(round(r2, 1)))
    
    
plt.legend( loc='lower right')    
plt.show()
```

# Test-Train Split

```python
# Train Test split 
x_train, x_test, y_train, y_test = train_test_split(x,y, 
                                     train_size = 0.5, 
                                     test_size = 0.5, 
                                     random_state = 42)

print(len(x_train))
print(len(x_test))
```

```python
#plot scatter plots of train and test data
plt.scatter(x_train,y_train, color='#71b3ef')
plt.scatter(x_test,y_test, color='#e00579')
```

# Models with varying Polynomial Degrees

```python
#train polynomial regression models with degrees: 1,2,8

degrees = [1,2,8]
colors=['b','y','m']

x_train=x_train.reshape(-1, 1)
y_train=y_train.reshape(-1, 1)
x_test=x_test.reshape(-1, 1)
y_test=y_test.reshape(-1, 1)


#plot scatter plots of train and test data
plt.scatter(x_train,y_train, color='#71b3ef')
plt.scatter(x_test,y_test, color='#e00579')


for i in range(len(degrees)):  
    poly = PolynomialFeatures(degree=degrees[i])
    poly_features = poly.fit_transform(x_train)
    model = LinearRegression()
    model.fit(poly_features, y_train)
    
    #predict y-values on train
    y_pred = model.predict(poly_features)
    
    #predict y-values on test
    poly_test_features = poly.fit_transform(x_test)
    y_test_pred = model.predict(poly_test_features)

    #Model Scores
    train_score= r2_score(y_train, y_pred) 
    test_score= r2_score(y_test, y_test_pred)
    
    #plot regression lines with labels
    sns.regplot(x=x_train,y=y_pred, order=degrees[i], scatter=False, line_kws={'color': colors[i], 'linewidth':2}, 
                ci=None, label='poly: '+str(degrees[i])+' train_score: '+str(round(train_score, 1))
                +' test_score: '+str(round(test_score, 1)))
    
    
plt.legend( loc='lower right')    
```

# Polynomial Model with varying Alpha

```python
#train polynomial regression model with varying alphas

alphas = [0,10,10000000000000000000]
colors=['m','y','b']

d=8

model_scores=[]

#plot scatter plots of train and test data
plt.scatter(x_train,y_train, color='#71b3ef')
plt.scatter(x_test,y_test, color='#e00579')

for i in range(len(alphas)):
    
    poly = PolynomialFeatures(degree=d)
    poly_features = poly.fit_transform(x_train.reshape(-1, 1))
    
    model_ridge = Ridge(alpha = alphas[i])
    model_ridge.fit(poly_features, y_train)
    
    #predict y-values on train
    y_pred = model_ridge.predict(poly_features)
    
    #predict y-values on test        
    poly_test_features = poly.fit_transform(x_test)
    y_test_pred = model_ridge.predict(poly_test_features)
    
    #model scores
    train_score= r2_score(y_train, y_pred) 
    test_score= r2_score(y_test, y_test_pred)

    #plot regression lines with labels
    sns.regplot(x=x_train,y=y_pred, order=d, scatter=False, line_kws={'color': colors[i], 'linewidth':2}, 
                ci=None, label='alpha:'+str(alphas[i])+' train_score: '+str(round(train_score, 1))
                +' test_score: '+str(round(test_score, 1)))
    
plt.legend( loc='lower right', title='polynomial degree= 8') 
```
