```python
#multivariate polynomial regression model with all twelve predictor variables

x1_train = df_train["area"]
y_train=df_train["price"]

x1_train = x1_train.to_numpy().reshape(-1,1)
y_train = y_train.to_numpy().reshape(-1,1)

x1_test = df_test["area"]
y_test=df_test["price"]

x1_test = x1_test.to_numpy().reshape(-1,1)
y_test = y_test.to_numpy().reshape(-1,1)

```

```python
#multivariate polynomial regression model with all twelve predictor variables


#train model
#train_scores=[]
#test_scores=[]

# checking models with polynomial degrees 1,2,3,4,5,6,7,8
degrees = [1]

for i in degrees:  
    poly = PolynomialFeatures(degree=1)
    poly_features = poly.fit_transform(x1_train)
    
    model_train = LinearRegression()
    model_train.fit(poly_features, y_train)
    y_train_pred = model_train.predict(poly_features)
    train_score= r2_score(y_train, y_train_pred)
    #test_score= model_train.score(x1_test, y_test)
    print("polynomial Degree:",i,"train score:", train_score)
```
